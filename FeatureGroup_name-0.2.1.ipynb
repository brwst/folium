Example FeatureGroup, using folium 0.2.1:

```python
import folium
from folium.map import *

print("folium: ", folium.__version__)

map = folium.Map()
feature_group = FeatureGroup(name='The name of my feature group')
feature_group.add_children(Marker([45,-30],popup=Popup('-30')))
feature_group.add_children(Marker([45, 30],popup=Popup('30')))
map.add_children(feature_group)
map.add_children(folium.map.LayerControl())
map
```

The FeatureGroup name argument is provided as 'The name of my feature group'.
