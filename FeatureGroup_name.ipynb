```python
import folium
from folium.map import *

print("folium: ", folium.__version__)

map = folium.Map()
feature_group = FeatureGroup(name='My feature group')
feature_group.add_children(Marker([45,-30],popup=Popup('-30')))
feature_group.add_children(Marker([45, 30],popup=Popup('30')))
map.add_children(feature_group)
map.add_children(folium.map.LayerControl())
map
```

The name of Marker FeatureGroup renders as **macro_element...**, not **My feature group**. The same behavior is seen in the FeatureGroup [here](https://ocefpaf.github.io/python4oceanographers/blog/2015/12/14/geopandas_folium/) as well.
